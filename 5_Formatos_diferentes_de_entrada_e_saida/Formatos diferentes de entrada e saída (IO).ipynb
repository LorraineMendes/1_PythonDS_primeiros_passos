## Módulo 1 - Alunas e alunos com Json - Pandas IO

```python
# Alteramos a versão da url para v1
# Adicionamos isso ao fim da url ?td=20&sexo=f

import pandas as pd
```

```python
url_fem = pd.read_json('https://servicodados.ibge.gov.br/api/v1/censos/nomes/ranking?qtd=200&sexo=f')
url_fem.head(10)
```

```python
url_masc = pd.read_json('https://servicodados.ibge.gov.br/api/v1/censos/nomes/ranking?qtd=200&sexo=m')
url_masc.head(10)
```

```python
print('Quantidade de nomes:{}' .format(str(len(url_fem) + len(url_masc))))
```

```python
frames = [url_fem, url_masc]
frames
```

```python
pd.concat(frames)
```

```python
pd.concat(frames)['nome'].to_frame() # transformando apenas a coluna nome em um dataframe
```

```python
nomes = pd.concat(frames)['nome'].to_frame()

nomes.sample(5) #traz nomes aleatórios
```

## Incluindo ID dos alunos

```python
import numpy as np
np.random.seed(123)
```

```python
total_alunos = len(nomes)
total_alunos
```

```python
nomes['id_alunos'] = np.random.permutation(total_alunos) +1
```

```python
nomes.sample(5)
```

```python
dominios = ['@dominiodoemail.com.br','@servicodoemail.com']
nomes['dominio']=np.random.choice(dominios, total_alunos)  #Elemento da lista aleatório
```

```python
nomes.sample(5)
```

```python
nomes['email'] = nomes.nome.str.cat(nomes.dominio).str.lower()
```

```python
nomes.sample(5)
```

```python
#escolher cursos de forma aleatório
#Precisamos saber quantas matrículas cada alunos tem
```

```python
todas_matriculas = []
x = np.random.rand(20) #quantidade de cursos
prob = x / sum(x)
```

```python
for index, row in nomes.iterrows(): #funçãoo que percorre linha a linha
    id = row.id_alunos #id do alunos
    matriculas = row.matriculas #Quantas matrículas esse aluno tem
    
    for i in range(matriculas):
        mat = [id, np.random.choice(cursos.index, p = prob)]
        todas_matriculas.append(mat)
        
matriculas = pd.DataFrame(todas_matriculas, columns = ['id_aluno', 'id_curso'])
```

```python
matriculas.head(10)
```

```python
#Quantos alunos tenho em cada curso

matriculas.groupby('id_curso').count().join(cursos['nome_do_curso'])
```

```python
matriculas.groupby('id_curso').count().join(cursos['nome_do_curso']).rename(columns ={'id_aluno':'Quantidade_de_alunos'})
```

```python
nomes.head(3)
```

```python
cursos.head(3)
```

```python
matriculas_por_curso = matriculas.groupby('id_curso').count().join(cursos['nome_do_curso']).rename(columns ={'id_aluno':'Quantidado_de_alunos'})
```
