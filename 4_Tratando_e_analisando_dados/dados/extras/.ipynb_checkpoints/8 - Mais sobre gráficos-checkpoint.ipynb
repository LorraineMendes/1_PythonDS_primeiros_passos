```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15,8))
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head()
```

```python
area = plt.figure() # Criando a área para o gráfico
```

```python
#adc 4 gráficos na figura
g1 = area.add_subplot(2, 2, 1) #gráfico 2 linhas, 2 colunas e o g1 fica na posição 1
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
g1.scatter(dados.Valor, dados.Area) #disperçao
g1.set_title('Valor X Área')

g2.hist(dados.Valor) #Histograma
g2.set_title('Histograma')

dados_g3 = dados.Valor.sample(100) #Amostra de tamanho 100
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')

grupo = dados.groupby('Tipo')['Valor']
label = grupo.mean().index #precisa dos rotolos e dos valores
valores = grupo.mean().values
g4.bar(label, valores)
g4.set_title('Valor Médio por Tipo')
```
